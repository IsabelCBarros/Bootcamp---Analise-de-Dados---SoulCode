# *8.1 Adição*

```python
# Adição
n1 = 5.887
n2 = 7.15
soma = n1 + n2

# Modos de apresentação
print(soma)
print('A soma é: ', round(soma,2))
print(f"A soma dos números é {soma:.2f}")
print(n1 + n2)
n1 + n2

# Verificação de tipos
print(type(soma))

# A função round() retorna um número de ponto flutuante que é uma versão arredondada do número especificado, com o número especificado de casas decimais .
# O número padrão de decimais é 0, o que significa que a função retornará o inteiro mais próximo.
# No exemplo, o número de casas decimais é 2 (soma, 2), o valor retornado será com 2 casas decimais (13.04).
```

##**31. FUNÇÃO RANGE - (permite especificar o inicio de uma sequência, o passo (ou pulo) e valor final.)**##

A função range() permite-nos especificar o início da sequência, o passo, e o valor final. O único parâmetro obrigatório é o que define quem será o último elemento da sequência.

A função range() retorna um conjunto de números sequenciais, que podem ser definidos de acordo com os parâmetros utilizados pela função. Por padrão, o intervalo gerado por ela começa pelo número zero e é incrementado um a um até atingir o ponto de parada.

Por gerar números sequenciais, uma de suas utilidades é servir de apoio para a construção da estrutura de repetição for, que permite executar uma instrução ou um bloco de código por várias vezes.

```python
# Função range com um argumento cria um lista com n valores inteiros de 0 a n-1
lista1 = list(range(5))
print(lista1)

# Função range com dois argumentos cria um lista com n valores inteiros do início informado no primeiro argumento até  ((n-1) - primeiro argumento informado) informado no segundo argumento
lista2 = list(range(1,5))
print(lista2)

# Função range com três argumentos cria um lista com definição de onde começar (primeiro argumento), onde parar (segundo argumento) e de qual tamanho de intervalo entre os valores (terceiro argumento)
lista3 = list(range(1,5,2))
print(lista3)
```

Primeiro, temos duas listas, uma chamada "pesos" e outra chamada "alturas". Essas listas contêm números que representam o peso e a altura de algumas pessoas.

Em seguida, o código começa a fazer alguns cálculos para cada pessoa. Ele usa um loop para percorrer cada elemento das listas de alturas e pesos.

Dentro do loop, ele pega a altura de uma pessoa e eleva ao quadrado, que é multiplicar a altura por ela mesma. Por exemplo, se a altura for 1.85, ele calcula 1.85 * 1.85.

Depois, ele armazena o resultado desse cálculo em uma nova lista chamada "alturas_quad". Essa lista vai guardar os valores das alturas elevadas ao quadrado para cada pessoa.

Em seguida, ele calcula o IMC (Índice de Massa Corporal) para cada pessoa. Para isso, ele divide o peso da pessoa pelo valor da altura ao quadrado que foi calculado anteriormente.

Ele armazena os resultados do IMC em outra lista chamada "imc".

Por fim, o código imprime a lista de IMCs calculados.

```python
# Utilizando o for com range
alturas = [1.85, 1.29, 1.75, 1.67, 1.82, 1.70]
alturas_quad = []
imc = []
for i in range(len(alturas)):
  alturas_quad.append(alturas[i]**2)
  imc.append(pesos[i]/alturas_quad[i])
print(f'IMC {imc}')
```

##**37. LOOP WHILE (O comando while faz com que um conjunto de instruções seja executado enquanto uma condição é atendida.)**##

O comando while faz com que um conjunto de instruções seja executado enquanto uma condição é atendida. Quando o resultado dessa condição passa a ser falso, a execução do loop é interrompida, como mostra o exemplo a seguir:

image.png

```python
contador = 0
while (contador < 5):
    print(contador)
    contador = contador + 1
```

Neste código, enquanto a variável contador, inicializada com 0, for menor do que 5, as instruções das linhas 3 e 4 serão executadas.

Observe que na linha 4 incrementamos o valor da variável contador, de forma que em algum momento seu valor igualará o número 5. Quando isso for verificado na linha 2, o laço será interrompido. Sem esse código, a condição de parada não será atingida, gerando o que é chamado de loop infinito. Evite que isso aconteça, pois leva ao congelamento e finalização da aplicação.
